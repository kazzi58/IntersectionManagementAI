```python
def SE(Path, d):
    #South-East
    if Path[1]==1:
        d.append(draw.ArcLine(30,-30,35,90,180,
                stroke='green', stroke_width=10,stroke_opacity=0.8, fill='none'))
        d.append(draw.Line(-5,-100,-5,-30,stroke='green', stroke_width=10,stroke_opacity=0.8, fill='none'))
        d.append(draw.Line(100,5,30,5,stroke='green', stroke_width=10,stroke_opacity=0.8, fill='none'))
    return d
#checked
#     d.append(draw.ArcLine(30,-30,35,90,180,
#                 stroke='#eeeeee', stroke_width=10,stroke_opacity=1, fill='none'))
```

```python
def EW(Path, d):
    #East-West
    if Path[2]==1:
        d.append(draw.Line(100,-15,-100,-15,stroke='green', stroke_width=10,stroke_opacity=0.8, fill='none'))
    return d
```

```python
def WE(Path, d):
    #West-East
    if Path[6]==1:
        d.append(draw.Line(100,15,-100,15,stroke='green', stroke_width=10,stroke_opacity=0.8, fill='none'))
    return d
```

```python
def SN(Path, d):
    #South-North
    if Path[0]==1:
        d.append(draw.Line(-15,-100,-15,100,stroke='green', stroke_width=10,stroke_opacity=0.8, fill='none'))
    return d
```

```python
def NS(Path, d):
    #North-South
    if Path[4]==1:
        d.append(draw.Line(15,-100,15,100,stroke='green', stroke_width=10,stroke_opacity=0.8, fill='none'))
    return d
```

```python
def drawNew(Path):
    dr = draw.Drawing(200, 200, origin='center')
    dr = DrawStreet(dr)
    dr = SE(Path,dr)
    dr = ES(Path,dr)
    dr = NE(Path,dr)
    dr = EN(Path,dr)
    dr = SW(Path,dr)
    dr = WS(Path,dr)
    dr = NW(Path,dr)
    dr = WN(Path,dr)
    dr = NS(Path,dr)
    dr = SN(Path,dr)
    dr = EW(Path,dr)
    dr = WE(Path,dr)
    
    dr.setPixelScale(3)  # Set number of pixels per geometry unit
    dr.setRenderSize(300,300)  # Alternative to setPixelScale
    dr.saveSvg('example.svg')
    dr.savePng('example.png')

    # Display in iPython notebook
    dr.rasterize()  # Display as PNG
    return dr  # Display as SVG
```
